```python
s1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s2 = pd.Series([10, 20, 30], index=['c', 'b', 'a'])
s3 = pd.Series([10, 20, 30, 40, 50, 60], index = ['a', 'a', 'a', 'b', 'c', 'c'])
s4 = pd.Series([10, 20], index=['a', 'b'])
s5 = pd.Series([10, 20, 30], index=['d', 'e', 'f'])
s1 + s2
s1 + s3
s3 + s4
s3 + s5
```

```python
s1 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
s2 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'cc', 'dd'])

s = s1 + s2
s.isna()
s.isnull()
s.notnull()
s[s.isnull()].index
```

```python
s1 = pd.Series([1, 2, 3, 4], index=['b', 'd', 'a', 'c'])
s2 = pd.Series([10, 20, 30, 40], index=['dd', 'c', 'bb', 'a'])

s = s1 + s2

s.dropna()
s.fillna(0)
s.fillna(s.mean())
```

```python
s.sort_values()                  # 오름차순
s.sort_values(ascending=False)   # 내림차순
s.sort_index()
```

```python
data = {
    '홍길동' : 95,
    '이순신' : 90,
    '이길동' : 45,
    '김둘리' : 72,
    '길동이' : 10
}
s = pd.Series(data)
s_stat = pd.Series([s.mean(), s.max(), s.min()] , index=['평균','최대','최소'])
print(s_stat)
print( s[ s > 80 ] )
```
