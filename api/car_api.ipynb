### 공공 데이터 포털
- https://www.data.go.kr/data/15059401/openapi.do

```python
import requests
import xml.etree.ElementTree as ET
from dotenv import load_dotenv
import os

load_dotenv()
service_key = os.getenv('public_service_key')
url = 'https://apis.data.go.kr/B553881/newRegistlnfoService_02/getnewRegistlnfoService02'

params = {
    'serviceKey' : service_key,
    'registYy' : '2026',
    'registMt' : '02',
    'vhctyAsortCode' : 1,
    'registGrcCode' : 1
}

response = requests.get(url, params=params)
# print( f"코드 : {response.status_code}" )
print(response.text) # 문자열 xml구조
# print( type(response.text) )

root = ET.fromstring(response.text)

for child in root.find('header'):
    print(child.tag, child.text)
for child in root.find('body'):
    print(child.tag, child.text)

print( root.findtext('body/dtaCo' ))
```
