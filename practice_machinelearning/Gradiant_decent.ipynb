```python
import numpy as np
import matplotlib.pyplot as plt

# 가장 단순한 손실 함수 : f(w) = w^2
# 최솟값은 w=0 일 때 (우리 눈엔 보이지만 모델은 모름)

def loss(w):
    return w ** 2

def gradient(w):
    return 2 * w # f(w) = w^2 의 미분 = 2w
```

```python
# 시작점, 학습률 설정
w = 10.0 # 임의의 시작점
lr = 1.1 # 학습률 (한 걸음의 크기)
epochs = 50 # 몇 번 이동할지

history_w = [w]
history_loss = [loss(w)]

for i in range(epochs):
    grad = gradient(w) # 현재 위치의 기울기
    w = w - lr * grad # 기울기 반대 방향으로 이동
    history_w.append(w)
    history_loss.append(loss(w))

print(f"최종 w : {w:.6f}") # 0에 가까워야 함
print(f"최종 loss : {loss(w):.6f}")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# 왼쪽 : w가 어떻게 변하는지
ax1.plot(history_w, color='blue')
ax1.axhline(0, color='red', linestyle='--', label='정답 w=0')
ax1.set_xlabel('epoch')
ax1.set_ylabel('w')
ax1.set_title('w의 변화')
ax1. legend()

# 오른쪽 : 손실이 어떻게 줄어드는지
ax2.plot(history_loss, color='orange')
ax2.set_xlabel('epoch')
ax2.set_ylabel('loss')
ax2.set_title('loss의 변화')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
lr_list = [0.1, 0.4, 0.9, 1.0, 1.1, 2.0]

for ax, lr in zip(axes.flatten(), lr_list):
    w = 10
    history = [w]

    for _ in range(30):
        grad = 2 * w
        w = w - lr * grad
        history.append(w)
        if abs(w) > 1e10: # 너무 커지면 중단
            break
    
    ax.plot(history)
    ax.axhline(0, color='red', linestyle='--')
    ax.set_title(f"lr = {lr}")
    ax.set_ylim(-50, 50)

plt.tight_layout()
plt.show()
```

```python
# w = 0.0
# 손실 함수 : f(w) = (w - 3)^2 <- 최솟값이 w=3인 함수
# lr = 0.1, epochs = 50

w = 8.0
lr = 0.1
epochs = 50

history_w = [w]
history_loss = [(w-3)**2]

for i in range(epochs):
    grad = 2 * w - 6
    w = w - lr * grad
    loss = (w-3)**2
    history_w.append(w)
    history_loss.append(loss)

print(f"최종 w : {w:.6f}")
print(f"최종 loss : {loss:.6f}")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(history_w, color='blue')
ax1.axhline(3, linestyle='--', color='red')
ax1.set_xlabel('epoch')
ax1.set_ylabel('w')

ax2.plot(history_loss, color='green')
```
