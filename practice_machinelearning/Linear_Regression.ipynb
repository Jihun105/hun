```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성
np.random.seed(42)
x = np.linspace(0, 10, 50) # 0부터 10까지를 동일 간격 50개로 나눈 배열
y = 0.1 * x + 0.5 + np.random.randn(50) * 0.1 # 노이즈 추가
# 진짜 관계 : w=0.1, b=0.5를 찾아야함

w = 0.0
b = 0.0
lr = 0.01
epochs = 1000

history_loss = [0]
for i in range(epochs):
    dw = -2 * np.mean( x * (y - (w*x + b )) )
    db = -2 * np.mean(y - (w*x + b))
    w = w - lr * dw
    b = b - lr * db
    y_hat = w * x + b
    loss = np.mean( (y - y_hat)**2 )
    history_loss.append(loss)

    if i % 100 == 0:
        print(loss)

print(f"최종 w : {w}, b : {b}")

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].scatter(x, y)
axes[0].plot(x, y_hat, color='red', label=f'y={w:.3f}x+{b:.3f}')
axes[1].plot(history_loss)
```
